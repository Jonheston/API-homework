```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
cities
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
```

```python
latitude = []
longitude = []
temperature = []
humidity = []
wind = []
cloudyness = []
counter = 1
passed_cities = []
country = []

for city in cities:
    try:
        weather = requests.get(url + "&q=" + city)
        weather_json = weather.json()
        latitude.append(weather_json["coord"]["lat"])
        longitude.append(weather_json["coord"]["lon"])
        temperature.append(weather_json["main"]["temp"])
        humidity.append(weather_json["main"]["humidity"])
        wind.append(weather_json["wind"]["speed"])
        cloudyness.append(weather_json["clouds"]["all"])
        passed_cities.append(city.title())
        country.append(weather_json["sys"]["country"])
        print(f"{counter} of {len(cities)}. {city} weather report is complete")
        counter = counter + 1
        
    except:
        print(f"{counter} of {len(cities)}. {city} weather report could not be found")
        counter = counter + 1

   
```

```python
# Create data frame and CSV
weather_data = pd.DataFrame({"City": passed_cities, "Latitude": latitude, "Longitude": longitude, "Country Code": country, "Humidity (%)": humidity, "Wind Speed (mph)": wind, "Cloudyness (%)": cloudyness})
weather_data
weather_data.to_csv("Weather Data.csv")
```
